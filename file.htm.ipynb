```python
import numpy as np
import matplotlib.pyplot as plt

r = np.arange(0, 10, 0.2)
theta = 2 * np.pi * r
fig, ax = plt.subplots(
  subplot_kw = {'projection': 'polar'} 
)
ax.plot(theta, r)
ax.set_rticks([0, 2.5, 5, 7.5, 10])
ax.grid(True)
plt.show()
```
